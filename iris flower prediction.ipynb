```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
df2=load_iris()
p=df2.data
q=df2.target
```

```python
p
```

```python
labels=['sepal length','sepal width','petal length','petal width']
label1=['type']
```

```python
d1=pd.DataFrame(p,columns=labels)
d1
```

```python
d2=pd.DataFrame(q,columns=label1)
```

```python
df=pd.concat([d1,d2],axis=1)
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
x=df.corr()
#obtaining correlation
plt.figure(figsize=(10,6))
sns.heatmap(x,annot=True)
#annot refers to annotation if we remove it will not show the values only plain graph shown
plt.title('correlation matrix')
plt.xlabel('features')
plt.ylabel('features')
plt.show()
```

```python
x=df.drop('type',axis=1)
y=df[['type']]
y=np.array(y)
y=y.ravel()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
m1=LogisticRegression()
```

```python
m1.fit(xtrain,ytrain)
```

```python
m1.score(xtrain,ytrain)
```

```python
m1.score(xtest,ytest)
```

```python
m2=DecisionTreeClassifier()
```

```python
m2.fit(xtrain,ytrain)
```

```python
m2.score(xtrain,ytrain)
```

```python
m2.score(xtest,ytest)
```

```python
joblib.dump(m2,"iris flower classification")
model=joblib.load("iris flower classification")
```

```python
def func():
    e1=e11.get()
    e2=e22.get()
    e3=e33.get()
    e4=e44.get()
    arr=np.array([e1,e2,e3,e4])
    x=model.predict(arr.reshape(1,-1))
    if x==0:
        s='Iris setosa'
    elif x==1:
        s='Iris versicolor'
    else:
        s='Iris virginica'
    p='The Flower is Classified as : '+s
    labelx.config(text=p)
    
    
m=Tk()
m.configure(bg="violet")
m.title("Iris Flower Classification")
m.geometry("750x500")
m.minsize(750,500)
m.maxsize(750,500)

label1=Label(m,text="Enter Sepal Length",bg="violet",fg="black",font=("Arial",15,"bold"))
label1.grid(row=0,column=0,pady=10,padx=50)
e11=Entry(m,width=20,font=("Arial",14),bg="pink",fg="black",borderwidth=2)
e11.grid(row=0,column=1,pady=10,padx=5)

label2=Label(m,text="Enter Sepal Width",bg="violet",fg="black",font=("Arial",15,"bold"))
label2.grid(row=1,column=0,pady=10,padx=50)
e22=Entry(m,width=20,font=("Arial",14),bg="pink",fg="black",borderwidth=2)
e22.grid(row=1,column=1,pady=10,padx=5)

label3=Label(m,text="Enter Petal Length",bg="violet",fg="black",font=("Arial",15,"bold"))
label3.grid(row=2,column=0,pady=10,padx=50)
e33=Entry(m,width=20,font=("Arial",14),bg="pink",fg="black",borderwidth=2)
e33.grid(row=2,column=1,pady=10,padx=5)

label4=Label(m,text="Enter Petal Width",bg="violet",fg="black",font=("Arial",15,"bold"))
label4.grid(row=3,column=0,pady=10,padx=50)
e44=Entry(m,width=20,font=("Arial",14),bg="pink",fg="black",borderwidth=2)
e44.grid(row=3,column=1,pady=10,padx=5)

labelx=Label(m,text="",bg="violet",fg="black",font=("Arial",15,"bold"))
labelx.grid(row=4,column=0,pady=10,padx=50)

btn=Button(text="Submit",bg="lightgreen",activebackground="pink",borderwidth=2,font=("Arial",15,"bold"),command=func)
btn.grid(row=5,column=0,columnspan=2,padx=340,pady=10)
m.mainloop()
    
```
